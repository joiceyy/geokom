1.4 Mengatur Axis pada Matplotlib

Mengatur nilai maksimum dan minimum pada Plot

Pengatuan nilai maksimum dan minimum dilakukan dengan perintah :


plt.xlim(min_boun, max_bound) atau plt.ylim(min_boun, max_bound)

```python
plt.hist(y)
plt.ylim(0,25)
```

```python
plt.hist(y)
plt.ylim(0,150)
plt.xticks([0,4,8],
           ['nol', 'empat', 'delapan'],
           rotation=0);
```

Mengatur Skala pada Axis dan membalikkannya

```python
plt.hist(y)
plt.ylim(0.3,150)
plt.xticks([0,4.5,8],
           ['nol', 'empat', 'delapan'],
           rotation=0);

plt.gca().set_yscale('log')
plt.gca().invert_yaxis()
```

5.1 Menggunakan Subplot

```python
# creating dummy data
x1 = [1,4,6,8,10]
y1 = [-2,2,4,6,8]

x2 = [2,4,6,8,9]
y2 = [9,8,6,-1,10]
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.scatter(x1,y1)
plt.title('grafik 1')

plt.subplot(1,2,2)
plt.plot(x1,y1,'pink')
plt.title('grafik 2')
```
